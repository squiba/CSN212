```python
from grahamscan import *
from giftalgo import *
import time
```

```python
no_of_points = 1000
points_on_hull=[]
graham_time = []
gift_time = []
get_time = time.perf_counter
```

```python
all_on_hull = generate_circle_points(10)

start = get_time()
sol1 = grahamscan_e(all_on_hull)
end = get_time()
graham_time.append(end-start)
points_on_hull.append(len(sol1))
print(end-start)
print(len(sol1))

start = get_time()
sol2 = giftwrap_e(all_on_hull)
end = get_time()
gift_time.append(end-start)
print(len(sol2))
print(end-start)
```

```python
points  = generate_random_array(no_of_points)

start = get_time()
sol = grahamscan_e(points)
end = get_time()
graham_time.append(end-start)
points_on_hull.append(len(sol))
print(end-start)
print(len(sol))

start = get_time()
sol = giftwrap_e(points)
end = get_time()
gift_time.append(end-start)
print(len(sol))
print(end-start)
```

```python
points  = generate_random_array(no_of_points)

start = get_time()
sol = grahamscan_e(points)
end = get_time()
graham_time.append(end-start)
points_on_hull.append(len(sol))
print(end-start)
print(len(sol))

start = get_time()
sol = giftwrap_e(points)
end = get_time()
gift_time.append(end-start)
print(len(sol))
print(end-start)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(points_on_hull,graham_time,'-ro')
plt.plot(points_on_hull,gift_time,'-bo')
plt.xlabel("number of points on hull")
plt.ylabel("running time")
plt.legend(['Graham_Scan','Gift_Wrapping'], loc='upper left')
plt.title("n=1000")
plt.savefig("different_points.png")
```

```python
import numpy as np
no_of_points = [100, 1000, 2000, 5000, 10000,20000,30000,40000]
Graham_time_taken = []
Gift_time_taken =[]
header = ['no_of_points','Graham_scan_time_taken','Gift_wrapping_time_taken']
for i in range(len(no_of_points)):
    points  = generate_random_array(no_of_points[i])
    
    start = get_time()
    sol = grahamscan_e(points)
    end = get_time()
    Graham_time_taken.append(end-start)
    
    start = get_time()
    sol = giftwrap_e(points)
    end = get_time()
    Gift_time_taken.append(end-start)
    
np.save("hull_time",(header,no_of_points,Graham_time_taken,Gift_time_taken))
plt.plot(no_of_points,Graham_time_taken,'-o')
plt.plot(no_of_points,Gift_time_taken,'-o')
plt.xlabel("Number of points")
plt.ylabel("Running Time")
plt.legend(['Graham_Scan','Gift_Wrapping'], loc='upper left')
plt.title("Comparision chart between Graham and Gift algorithms with randomly generated points on a square grid")
plt.savefig("Comparision_Chart_new_2.png")
```
